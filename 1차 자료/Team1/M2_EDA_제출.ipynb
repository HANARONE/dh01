# 데이터 불러오기

```python
df_sg_train = pd.read_csv('/content/drive/MyDrive/hana1/Team/final/semi_good_train.csv', index_col = 0)
df_sg_train.shape
```

```python
df_sp_train = pd.read_csv('/content/drive/MyDrive/hana1/Team/final/semi_poor_train.csv', index_col = 0)
df_sp_train.shape
```

# EDA

## Semi_Good

```python
analyze_report = sv.analyze(df_sg_train)
# display & save the report
analyze_report.show_html('report_semi_good.html')
```

## Semi_Good <-> Semi_Poor

```python
compare_report = sv.compare([df_sg_train, 'Semi_Good'], [df_sp_train, 'Semi_Poor'])
# display & save the report
compare_report.show_html('report_semi_good_poor.html')
```
