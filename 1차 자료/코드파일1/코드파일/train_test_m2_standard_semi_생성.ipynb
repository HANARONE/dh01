# 데이터 불러오기

```python
df_standard_train = pd.read_csv('/content/drive/MyDrive/hana1/project1/data/drop_standard_train.csv', index_col = 0)
df_standard_train.shape
```

```python
df_standard_test = pd.read_csv('/content/drive/MyDrive/hana1/project1/data/drop_standard_test.csv', index_col = 0)
df_standard_test.shape
```

# 데이터 분할

```python
# X, y 분할
X_train = df_standard_train.drop('Credit_Score', axis = 1)
X_test = df_standard_test.drop('Credit_Score', axis = 1)

print("X_train 데이터: ", X_train.shape)
print("X_test 데이터: ", X_test.shape)
```

# Model 2 생성

## K-means

### train 적용

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2, random_state=42)
km.fit(X_train)
print(km.labels_)
print(np.unique(km.labels_, return_counts=True))
```

```python
df_X_train = X_train.copy()
df_X_train['Cluster'] = km.labels_
df_X_train
```

```python
clusters_train = df_X_train.groupby('Cluster')

df_clusters_mean_train = pd.DataFrame(clusters_train.mean())
df_clusters_mean_train.T
```

### test 적용

```python
km.fit(X_test)
print(km.labels_)
print(np.unique(km.labels_, return_counts=True))
```

```python
df_X_test = X_test.copy()
df_X_test['Cluster'] = km.labels_
df_X_test
```

```python
clusters_test = df_X_test.groupby('Cluster')
```

```python
df_clusters_mean_test = pd.DataFrame(clusters_test.mean())
df_clusters_mean_test.T
```
