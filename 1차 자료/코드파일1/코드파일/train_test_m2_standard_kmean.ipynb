```python
drop_standard_train = pd.read_csv('/content/drive/MyDrive/hana1/project1/data/drop_standard_train.csv', index_col = 0) # 20*28
drop_standard_test = pd.read_csv('/content/drive/MyDrive/hana1/project1/data/drop_standard_test.csv', index_col = 0) # 10 rows × 23
drop_standard_train.head()
drop_standard_test.head()
```

```python
drop_standard_train = drop_standard_train.drop('Credit_Score', axis=1)
drop_standard_test = drop_standard_test.drop('Credit_Score', axis=1)
```

# K-Means

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans


# K-Means 알고리즘을 사용하여 클러스터링
km = KMeans(n_clusters=2, random_state=42)
km.fit(drop_standard_train)

# 클러스터링 결과를 데이터프레임에 추가
drop_standard_train['Cluster'] = km.labels_

# 결과 확인
print(drop_standard_train.head())
```

```python
print(km.labels_)
print(np.unique(km.labels_,return_counts=True))
```

```python
drop_standard_train_copy = drop_standard_train.copy()
drop_standard_train_copy['Cluster'] = km.labels_
print(drop_standard_train_copy)
```

```python
drop_train_clusters = drop_standard_train_copy.groupby('Cluster')
```

```python
drop_standard_train_copy_mean = pd.DataFrame(drop_train_clusters.mean())
drop_standard_train_copy_mean.T
```

### test

```python
km.fit(drop_standard_test)

# 클러스터링 결과를 데이터프레임에 추가
drop_standard_test['Cluster'] = km.labels_

# 결과 확인
print(drop_standard_test.head())
```

```python
print(km.labels_)
print(np.unique(km.labels_,return_counts=True))
```

```python
drop_standard_test_copy = drop_standard_test.copy()
drop_standard_test_copy['Cluster'] = km.labels_
print(drop_standard_test_copy)
```

```python
drop_test_clusters = drop_standard_test_copy.groupby('Cluster')
```

```python
drop_standard_test_copy_mean = pd.DataFrame(drop_test_clusters.mean())
drop_standard_test_copy_mean.T
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 시각화 1 good, 0 poor
sns.scatterplot(data=drop_standard_train, x='Age', y='Annual_Income', hue='Cluster', palette='viridis')
plt.xlabel('Age')
plt.ylabel('Annual_Income')
plt.title('K-Means Clustering')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 시각화 1 good, 0 poor
sns.scatterplot(data=drop_standard_train, x='Annual_Income', y='Outstanding_Debt', hue='Cluster', palette='viridis')
plt.xlabel('Annual_Income')
plt.ylabel('Outstanding_Debt')
plt.title('K-Means Clustering')
plt.show()
```

## 실루엣점수

```python
# 실루엣 점수
from sklearn.metrics import silhouette_score
print(silhouette_score(drop_standard_train, drop_standard_train_copy['Cluster'], metric='euclidean'))
```

```python
# 실루엣 점수
from sklearn.metrics import silhouette_score
print(silhouette_score(drop_standard_test, drop_standard_test_copy['Cluster'], metric='euclidean'))
```
